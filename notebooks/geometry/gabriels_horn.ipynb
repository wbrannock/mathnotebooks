```python
import numpy as np
import matplotlib.pyplot as plt

# Domain from x = 1 to x = 10 (approaches ∞ visually)
x = np.linspace(1, 10, 1000)
y = 1 / x**2

# Create plot
plt.figure(figsize=(10, 5))

# Plot the curves y = 1/x^2 and y = -1/x^2
plt.plot(x, y, 'r', label=r'$y = \frac{1}{x^2}$')
plt.plot(x, -y, 'r', label=r'$y = -\frac{1}{x^2}$')

# Fill the area between the curves
plt.fill_between(x, -y, y, color='orange', alpha=0.6, label='Enclosed area')

# Plot settings
plt.title("2D Analog of Gabriel's Horn", fontsize=16)
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(1, color='black', linewidth=0.5, linestyle='--')
plt.axvline(10, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.ylim(-1.1, 1.1)
plt.xlim(0.8, 10.2)
#Add text showing area is finite
plt.text(5, 0.5, r'Area = $\int_1^{\infty} \frac{2}{x^2} dx = 2$', fontsize=14, color='blue', ha='center')
#Add text showing perimeter is infinite
plt.text(5, -0.7, r'Perimeter = $\int_1^{\infty} 2\sqrt{1 + \left(-\frac{2}{x^3}\right)^2} dx = \infty$', fontsize=14, color='blue', ha='center')
plt.show()
```

```python


# Define x range
x = np.linspace(1, 25, 400)
y = 1 / x

# Create a meshgrid for surface of revolution
theta = np.linspace(0, 2 * np.pi, 100)
X, Theta = np.meshgrid(x, theta)
Y = (1 / X) * np.cos(Theta)
Z = (1 / X) * np.sin(Theta)

# Plotting
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, linewidth=0, antialiased=True)

# Plot settings
ax.set_title("3D Gabriel's Horn: Surface of Revolution of y = 1/x", fontsize=14)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Set viewing angle for better perspective
ax.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()
```
