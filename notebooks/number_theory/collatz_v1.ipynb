```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Basic Collatz sequence implementation
step = 0
number = 9

while number > 1:
    if number % 2 == 0:
        number = number // 2
    else:
        number = 3 * number + 1
    step += 1

print(step)
```

```python


# Function to plot the Collatz sequence for a given starting number
def plot_collatz_for_number(n):
    x = [n]
    y = [0]
    step = 0
    while n > 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        step += 1
        x.append(n)
        y.append(step)
    plt.figure(figsize=(10, 6))
    plt.plot(y, x, marker='o', linestyle='-', markersize=5)
    plt.title(f'Collatz Conjecture Steps for Starting Number {x[0]}')
    plt.xlabel('Steps')
    plt.ylabel('Number')
    #Label max y value
    plt.text(y[np.argmax(x)], max(x), f'Max: {max(x)}', fontsize=10, verticalalignment='bottom', horizontalalignment='right')
    #make x axis go in integer steps
    plt.xticks(np.arange(0, step + 1, 1))
    plt.grid(True)
    plt.show()


plot_collatz_for_number(9)
```

```python
# Function to plot the number of steps to reach 1 for all starting numbers up to n
def plot_collatz(n):
    x = np.arange(1, n + 1)
    y = np.zeros(n)
    for i in range(1, n + 1):
        number = i
        step = 0
        while number > 1:
            if number % 2 == 0:
                number = number // 2
            else:
                number = 3 * number + 1
            step += 1
        y[i - 1] = step
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, marker='o', linestyle='-', markersize=3)
    plt.title('Collatz Conjecture Steps')
    plt.xlabel('Starting Number')
    plt.ylabel('Number of Steps to Reach 1')
    #Overlay line with rolling average number of steps for all starting numbers
    window_size = 100
    y_avg = np.convolve(y, np.ones(window_size)/window_size, mode='valid')
    plt.plot(x[window_size-1:], y_avg, color='red', label='Rolling Average Steps (window size 1000)')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_collatz(2500)
```
